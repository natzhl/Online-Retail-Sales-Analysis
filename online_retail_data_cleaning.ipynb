```python
import kagglehub

# Загрузка датафрейма
path = kagglehub.dataset_download("ulrikthygepedersen/online-retail-dataset")
```

# Очистка и подготовка данных

```python
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join(path, "online_retail.csv"))
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum() # проверяем нули
```

```python
df = df[df["Quantity"] > 0] # удаляем возвраты
df = df[df["UnitPrice"] > 0] # удаляем нулевые и отрицательные цены
```

```python
df = df.dropna(subset=["CustomerID"]) # удаляем строки без id покупателя
df['CustomerID'] = df['CustomerID'].astype(int)
df.info()
```

```python
df.isna().sum() # проверяем нули
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```
